# MNIST Handwritten Digit Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist

# Load MNIST data
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize the data
x_train = x_train / 255.0
x_test = x_test / 255.0

# Class names for the digits 0-9
class_names = [str(i) for i in range(10)]

# Visualize the first 10 images
plt.figure(figsize=(10, 10))
for i in range(10):
    plt.subplot(5, 5, i + 1)
    plt.grid(False)
    plt.imshow(x_train[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[y_train[i]])
    plt.tight_layout()
plt.show()

# Define activation functions and their derivatives
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def d_relu(x):
    return np.where(x > 0, 1, 0)

def d_sigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))

def d_tanh(x):
    return 1 - np.square(np.tanh(x))

# Test the activation functions
x = np.linspace(-10, 10, 100)
plt.figure(figsize=(10, 6))

plt.plot(x, sigmoid(x), label="Sigmoid")
plt.plot(x, relu(x), label="ReLU")
plt.plot(x, np.tanh(x), label="Tanh")

plt.title("Activation Functions")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.grid()
plt.show()

# Example of implementing forward propagation and initializing weights
# Simplified for demonstration (doesn't train on MNIST yet)
hiddenlayers = 2
hiddennodes = 64
inputnode = 784  # 28x28 flattened image
outputnode = 10  # 10 digits

# Initialize weights and biases randomly
W = {1: np.random.randn(hiddennodes, inputnode) * 0.01,
     2: np.random.randn(outputnode, hiddennodes) * 0.01}
b = {1: np.zeros((hiddennodes, 1)),
     2: np.zeros((outputnode, 1))}

# Forward propagation example with ReLU activation
def forward_prop(x_sample):
    x_flat = x_sample.reshape(-1, 1)  # Flatten the image
    z1 = np.dot(W[1], x_flat) + b[1]
    a1 = relu(z1)
    z2 = np.dot(W[2], a1) + b[2]
    yhat = sigmoid(z2)  # Output layer with sigmoid
    return yhat

# Test forward propagation on a single MNIST image
yhat_sample = forward_prop(x_train[0])
print("Output probabilities for the first image:\n", yhat_sample)
```
